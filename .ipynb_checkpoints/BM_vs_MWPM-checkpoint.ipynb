```python
import numpy as np
import stim
import pymatching
import matplotlib.pyplot as plt

from beliefmatching import BeliefMatching
```

```python
def count_logical_errors(circuit: stim.Circuit, num_shots: int, circuit_drifted: stim.Circuit = None) -> int:

    if circuit_drifted != None:
        # sampler = circuit_drifted.compile_detector_sampler()
        sampler = circuit.compile_detector_sampler()
    else:
        sampler = circuit.compile_detector_sampler()
    detection_events, observable_flips = sampler.sample(num_shots, separate_observables=True)

    # Configure a decoder using the circuit.
    if circuit_drifted != None:
        detector_error_model = circuit_drifted.detector_error_model(decompose_errors=True) # use error model from original circuit
    else:
        detector_error_model = circuit.detector_error_model(decompose_errors=True)
    matcher = pymatching.Matching.from_detector_error_model(detector_error_model)

    # Run the decoder.
    predictions = matcher.decode_batch(detection_events)

    # Count the mistakes.
    num_errors = 0
    for shot in range(num_shots):
        actual_for_shot = observable_flips[shot]
        predicted_for_shot = predictions[shot]
        if not np.array_equal(actual_for_shot, predicted_for_shot):
            num_errors += 1
    return num_errors

def count_logical_errors_bf(circuit: stim.Circuit, num_shots: int, circuit_drifted: stim.Circuit = None) -> int:

    sampler = circuit.compile_detector_sampler()
    shots, observables = sampler.sample(num_shots, separate_observables=True)

    bm = BeliefMatching(circuit, max_bp_iters=20)

    # Run the decoder.
    # predictions = matcher.decode_batch(detection_events)
    predicted_observables = bm.decode_batch(shots)

    num_errors = np.sum(np.any(predicted_observables != observables, axis=1))

    # Count the mistakes.
    # num_errors = 0
    # for shot in range(num_shots):
    #     actual_for_shot = observable_flips[shot]
    #     predicted_for_shot = predictions[shot]
    #     if not np.array_equal(actual_for_shot, predicted_for_shot):
    #         num_errors += 1
    return num_errors
```

```python
def get_drift_factor(t, T_mean=14.8,  T_std=0.25):
    # Convert T_mean and T_std to the parameters of the underlying normal distribution
    mu = np.log(T_mean) - 0.5 * T_std**2
    sigma = T_std
    
    # Sample T_drift from the log-normal distribution
    T_drift = np.random.lognormal(mean=mu, sigma=sigma)
    return (10 ** (t / T_drift))

num_gates = 4
t = 2
drift_factors = [get_drift_factor(t) for _ in range(num_gates)]
print(drift_factors)
```

```python
num_shots = 10_000

for d in [3, 5, 7]:
# for d in [3, 5]:
    print("d = ", d)
    xs = []
    ys_original = []
    ys_drifted = []
    for p in [0.006, 0.007, 0.008, 0.009, 0.01]:
        print("p = ", p)
        circuit=stim.Circuit.generated(
                    rounds=d,
                    distance=d,
                    after_clifford_depolarization=p,
                    after_reset_flip_probability=p,
                    before_measure_flip_probability=p,
                    before_round_data_depolarization=p,
                    code_task=f'surface_code:rotated_memory_x',
                )
        circuit_drifted=stim.Circuit.generated(
                    rounds=d,
                    distance=d,
                    after_clifford_depolarization=p * drift_factors[0],
                    after_reset_flip_probability=p * drift_factors[1],
                    before_measure_flip_probability=p * drift_factors[2],
                    before_round_data_depolarization=p * drift_factors[3],
                    code_task=f'surface_code:rotated_memory_x',
                )

        num_errors_original = count_logical_errors(circuit, num_shots)
        num_errors_drifted = count_logical_errors_bf(circuit, num_shots)
        xs.append(p)
        ys_original.append(num_errors_original / num_shots)
        ys_drifted.append(num_errors_drifted / num_shots)

    # Plot original data with solid lines and circle markers.
    plt.plot(xs, ys_original, marker='o', linestyle='-', label=f"MWPM, d={d}")
    # Plot drifted data with dashed lines and square markers.
    plt.plot(xs, ys_drifted, marker='s', linestyle='--', label=f"Belief-Matching, d={d}")

plt.loglog()
plt.grid()
plt.xlabel("Physical Error Rate")
plt.ylabel("Logical Error Rate per Shot")
plt.legend()
plt.savefig("comparison_orig.png")
plt.show()

print("MWPM logical error rates:", ys_original)
print("BM logical error rates:", ys_drifted)
```
